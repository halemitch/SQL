```python
url = "https://yfapi.net/v1/finance/trending/US"

headers = {
    'x-api-key': apikey,
    }

response = requests.get(url, headers=headers, params=querystring)

print(response.text)
```

```python

#i only got the top 5 because I was having limit issues
df6 = pd.DataFrame(response.json()['finance']['result'][0]['quotes'][0:5])

trending_sym = []
for each in df6['symbol']:
    trending_sym.append(each)
trending_sym
```

```python

url = "https://yfapi.net/v6/finance/quote"

querystring = {"symbols":",".join(trending_sym)}

headers = {
    'x-api-key': apikey
    }

response = requests.request("GET", url, headers=headers, params=querystring)

print(response.text)
```

```python
response = requests.get(url, headers=headers, params=querystring)
df4 = pd.DataFrame(response.json()['quoteResponse']['result'])
df4.head()
```

```python
trending = df4[['regularMarketPrice', 'longName', 'symbol']]
trending
```

```python
list_curprice = []
for each in trending_sym:
    url = "https://yfapi.net/v11/finance/quoteSummary/" + each

    querystring = {"symbols":stock, 'modules': 'financialData'}

    headers = {
        'x-api-key': apikey,
        }

    response = requests.request("GET", url, headers=headers, params=querystring)
    response = requests.get(url, headers=headers, params=querystring)
    temp = pd.DataFrame(response.json()['quoteSummary']['result'][0])
    list_curprice.append(temp['financialData']['currentPrice']['raw'])
    print(list_curprice)
```

```python
list_52high = []
for each in trending_sym:

    url = "https://yfapi.net/v11/finance/quoteSummary/" + each

    querystring = {"symbols":stock, 'modules': 'summaryDetail'}

    headers = {
        'x-api-key': apikey,
        }

    response = requests.request("GET", url, headers=headers, params=querystring)
    response = requests.get(url, headers=headers, params=querystring)
    temp = pd.DataFrame(response.json()['quoteSummary']['result'][0])
    list_52high.append(temp['summaryDetail']['fiftyTwoWeekHigh']['raw'])
    print(list_52high)
```

```python
list_52low = []
for each in trending_sym:

    url = "https://yfapi.net/v11/finance/quoteSummary/" + each

    querystring = {"symbols":stock, 'modules': 'summaryDetail'}

    headers = {
        'x-api-key': apikey,
        }

    response = requests.request("GET", url, headers=headers, params=querystring)
    response = requests.get(url, headers=headers, params=querystring)
    temp = pd.DataFrame(response.json()['quoteSummary']['result'][0])
    list_52low.append(temp['summaryDetail']['fiftyTwoWeekLow']['raw'])
    print(list_52low)
```

```python
trending['52.Week.High'] = list_52high
trending['52.Week.Low'] = list_52low
trending['Current Price'] = list_curprice
trending
```
